```python
from minisom import MiniSom
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
```

```python
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
```

```python
som = MiniSom(x=10, y=10, input_len=4, sigma=1.0, learning_rate=0.5)
som.random_weights_init(X_normalized)
```

```python
print("Training SOM...")
som.train_random(data=X_normalized, num_iteration=1000)
print("Training complete.")
```

```python
plt.figure(figsize=(10, 10))
frequencies = np.zeros((10, 10))
```

```python
for x in X_normalized:
    w = som.winner(x)
    frequencies[w[0], w[1]] += 1

plt.imshow(frequencies.T, cmap='Blues')
plt.colorbar(label='Hits')
plt.title('SOM Hit Map')
plt.xlabel('SOM X')
plt.ylabel('SOM Y')
plt.grid()
plt.show()
```
